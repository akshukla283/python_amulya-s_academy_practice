```python
#### Now Adding insertion operation 
# 1. Pre-order - Root node --> Left sub tree --> right sub tree.
# 2. In-order - Left sub tree --> root node --> right sub tree.
# 3. Post-order - Left sub tree --> right sub tree --> root node

class BinarySearchTree:
    
    def __init__(self, key):
        self.key = key
        self.lchild = None
        self.rchild = None
    
    def insert(self, data):
        
        if self.key is None:
            self.key = data
            return
        
        if self.key == data :
            return
        
        if self.key > data:
            if self.lchild:
                self.lchild.insert(data)
            else:
                self.lchild = BinarySearchTree(data)
                
        else:
            if self.rchild:
                self.rchild.insert(data)
            else:
                self.rchild = BinarySearchTree(data)
                
    def search(self, data):
        
        if self.key == data:
            print("Node is found!")
            return
        
        if self.key > data:
            if self.lchild:
                self.lchild.search(data)
            else:
                print("Node is not present in the Tree!")
        
        else:
            if self.rchild:
                self.rchild.search(data)
            else:
                print("Node is not present in the Tree!")
                
    # Now we will write traversal operation
    
    def pre_order(self):   # pre-order -: root node --> left sub tree --> right sub tree
        
        print(self.key, end = " ")
        
        if self.lchild:
            self.lchild.pre_order()
            
        if self.rchild:
            self.rchild.pre_order()
            
    def in_order(self):  # in_order -: left sub tree --> root node --> rightsub tree
        
        if self.lchild:
            self.lchild.in_order()
        
        print(self.key, end = " ")
        
        if self.rchild:
            self.rchild.in_order()
            
    def post_order(self):  # post_order -: left sub tree --> right sub tree --> root node
        
        if self.lchild:
            self.lchild.post_order()
            
        if self.rchild:
            self.rchild.post_order()
            
        print(self.key, end = " ")
        
        
        
        
root = BinarySearchTree(10)

list1 = [20, 4, 30, 4, 1, 5, 6]
for i in list1:
    root.insert(i)
    
print("Pre_order: ")
root.pre_order()
print()
print("In_order : ")
root.in_order()
print()
print("Post_order : ")
root.post_order()
            
```

```python
#### Adding Deletion operation


class BinarySearchTree:
    
    def __init__(self, key):
        self.key = key
        self.lchild = None
        self.rchild = None
        
    def insert(self, data):
        
        if self.key is None:
            self.key = data
            return 
        
        if self.key == data:
            return 
        
        if self.key > data :
            if self.lchild:
                self.lchild.insert(data)
            else:
                self.lchild = BinarySearchTree(data)
        else:
            if self.rchild:
                self.rchild.insert(data)
            else:
                self.rchild = BinarySearchTree(data)
                
    # search operation
    def search(self, data):
        if self.key == data:
            print("Node is found!")
            return
        
        if self.key > data:
            if self.lchild:
                self.lchild.search(data)
            else:
                print("Given node is not present in the Tree!")
                
        if self.key <data:
            if self.rchild:
                self.rchild.search(data)
            else:
                print("Given node  is not present in the Tree!")
                
    # traversal operation
    
    def pre_order(self):
        print(self.key, end = " ")
        
        if self.lchild:
            self.lchild.pre_order()
            
        if self.rchild:
            self.rchild.pre_order()
            
    def in_order(self):
        if self.lchild:
            self.lchild.in_order()
            
        print(self.key, end = " ")
        
        if self.rchild:
            self.rchild.in_order()
            
    def post_order(self):
        if self.lchild:
            self.lchild.post_order()
            
        if self.rchild:
            self.rchild.post_order()
            
        print(self.key, end = " ")
        
    # Deletion operation
    def delete(self, data, curr):
        if self.key is None:
            print("Tree is empty!")
            return
        # if node is present then we need to find the current node
        
        if data < self.key:
            if self.lchild:
                self.lchild = self.lchild.delete(data, curr)
            else:
                print("Given node is not present in the Tree!")
                
        elif data > self.key:
            if self.rchild:
                self.rchild = self.rchild.delete(data, curr)
            else:
                print("Given node is not present in the Tree!")
                
        else:  # means data == self.key
            if self.lchild is None:        # means when one node rchild is present
                
                temp = self.rchild   
                # now check whether node is root node
                if data == curr:   # then we need to copy that node as root node
                    
                    self.key = temp.key
                    self.lchild = temp.lchild
                    self.rchild = temp.rchild
                    
                    temp = None
                    return 
                
                self = None
                return temp
            
            if self.rchild is None:
                temp = self.lchild
                if data == curr:  # then we need to copy that node as root node
                    self.key = temp.key
                    self.lchild = temp.lchild
                    self.rchild = temp.rchild
                    temp = None
                    return 
                
                self = None
                return temp
            # now is above both the condition false means node have two child node
            # then we can replace deleted node with either highest value from lchild or lowest 
            # value from rchild here we are going to take rchild lowest value
            
            node = self.rchild
            while self.lchild:  # In rchild it will search for the lowest lchild value 
                node = self.lchild
                
            # now we have lowest value from rchild node
            
            self.key = node.key
            self.rchild = self.rchild.delete(node.key, curr)
            
        return self
        
    def min_node(self):
        
        current = self
        while current.lchild:
            current = current.lchild
            
        # now we have leftmost value in current node
        print("Node with smallest key is : ", current.key)
        
    def max_node(self):
        current = self
        while current.rchild:
            current = current.rchild
            
        print("Node with largest key is : ", current.key)
        
        
def count(node):
    
    if node is None:
        return 0
    
    return 1 + count(node.lchild) + count(node.rchild)


root = BinarySearchTree(10)
list1 = [6, 3,7,8]
for i in list1:
    root.insert(i)

print("Pre-Order : ")
root.pre_order()

#root.delete(3, 10)

print()
print("Pre-Order : ")
root.pre_order()

print()
root.min_node()
root.max_node()
            
            
                    
```
