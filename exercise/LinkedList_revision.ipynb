```python
###  1. Singly Linked List

class Node:
    def __init__(self, data):
        self.data = data
        self.ref = None
        
        
class Singly_LinkedList:
    def __init__(self):
        self.head = None
        
    def print_LL(self):
        if self.head is None:
            print("Linked List is empty!")
        else:
            n = self.head
            while n is not None:
                print(n.data, "--->", end = " ")
                n = n.ref
                
    def add_empty(self, data):
        if self.head is None:
            new_node = Node(data)
            self.head = new_node
        else:
            print("Linked List is not empty!")
            
    def add_begin(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.ref = self.head
            self.head = new_node
            
    def add_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            n = self.head 
            while n.ref is not None:
                n = n.ref
                
            n.ref = new_node
            
    def add_after(self, data, x):
        n = self.head
        while n is not None:
            if n.data == x:
                break
            n = n.ref
        if n is None:
            print("Given node is not present is the Linked List!")
        else:
            new_node = Node(data) 
            new_node.ref = n.ref
            n.ref = new_node  # whether it may be last node will not affected 
            
    def add_before(self, data, x):
        
        if self.head is None:
            print("Can't add linked list is empty!")
            return
        
        # we are adding before so we need to check whether there is only one node
        if self.head.data == x:
            new_node = Node(data)
            new_node.ref = self.head
            self.head = new_node
            return
        
        n = self.head
        while n is not None:
            if n.ref.data == x:
                break
            n = n.ref
            
        if n is None:
            print("Given node is not present in the linked list!")
            
        else:
            new_node = Node(data)
            new_node.ref = n.ref
            n.ref = new_node
            
    def delete_begin(self):
        if self.head is None:
            print("Can't delete linked list is already empty!")
        else:
            self.head = self.head.ref
            
    def delete_end(self):
        if self.head is None:
            print("Can't delete linked list is already empty!")
        else:
            n = self.head
            while n.ref.ref is not None:
                n = n.ref
            n.ref = None
            
    def delete_by_value(self, x):
        if self.head is None:
            print("can't deelete LL is empty!")
            return # we use return so that after printig control should come out of the loop.
        # now chaeck wether the node is first node or not 
        if x == self.head.data:
            self.head = self.head.ref
            return  # if we don't want write else part then write 'return' or elif elif esle
        n = self.head  # using this we will search the head of the previous node
        while n.ref is not None:
            if x == n.ref.data:
                break
            n = n.ref
        #now we are here by two condition one is when while loop satisfied and second is when
        # nested if inside while loop satisfied and 'break' statement works.
        if n.ref is None:
            print("Node is not Found!")
        else:
            n.ref = n.ref.ref
            
                
            
        
            
            
        
                
                
                
sll1 = Singly_LinkedList()
sll1.add_empty(20)
# sll1.add_empty(20)
# sll1.add_begin(20)
# sll1.add_begin(30)
#sll1.add_begin(40)
#sll1.add_end(20)
#sll1.add_end(40)
# sll1.add_end(50)
#sll1.add_after(44, 20)
#sll1.add_before(100, 40)
sll1.delete_begin()


sll1.print_LL()
    
```

```python
### Doubly Linked List

class Node:
    def __init__(self, data):
        self.data = data
        self.nref = None
        self.pref = None
        
class Doubly_LinkedList:
    def __init__(self):
        self.head = None
        
    def print_DLL(self):
        if self.head is None:
            print("Linked List is empty!")
            
        else:
            n = self.head 
            while n is not None:
                print(n.data , "--->", end = " ")
                n = n.nref
                
    def print_reverse(self):
        if self.head is None:
            print("Linked List is empty!")
            
        else:
            n = self.head 
            while n.nref is not None:
                n = n.nref
            while n is not None:
                print(n.data, "--->", end= " ")
                n = n.pref
            
    def add_empty(self, data):
        if self.head is None:
            new_node = Node(data)
            self.head = new_node
        else:
            print("Linked List is not empty")
            
    def add_begin(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head  = new_node
        else:
            new_node.nref = self.head
            self.head.pref = new_node
            self.head = new_node
            
    def add_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            n = self.head
            while n.ref is not None:
                n = n.ref
            n.nref = new_node
            new_node.pref = n
    
    def add_after(self, data, x):
        if self.head is None:
            print("Linked List is empty!")
            
        else:
            n = self.head
            while n is not None:
                if n.data == x:
                    break
                n = n.nref
                
            if n is None:
                print("Given Node is not present in Linked List!")
            else:
                new_node = Node(data)
                new_node.nref = n.nref
                new_node.pref = n
                if n.nref is not None : # means this node may be from first to before last 
                    n.nref.pref = new_node
                    
                n.nref = new_node
                
    def add_before(self, data, x):
        if self.head is None:
            print("Linked list is empty!")
        else:
            n = self.head
            while n is not None:
                if n.nref.data = x:
                    break
                n = n.ref
                
            if n is None:
                print("Given node is not present!")
            else:
                new_node = Node(data)
                new_node.nref = n.nref
                new_node.pref = n
                if n.pref is not None:
                    n.nref.pref = new_node
                else:
                    self.head = new_node
                    
                n.nref = new_node
                
    def delete_begin(self):
        if self.head is None:
            print("DLL is empty can't delete!")
            return
        if self.head.nref is None:
            self.head = None
            print("DLL is empty after deleting the node!")
        else:
            self.head = self.head.nref
            self.pref = None
            
    def delete_end(self):
        if self.head is None:
            print("DLL is empty can't delete!")
            return
        if self.head.nref is None:
            self.head = None
            print("DLL is empty after deleting the node!")
        else:
            n = self.head
            while n is not None:
                n = n.nref
            n.pref.nref = None
    
    def delete_by_value(self, x):
        if self.head is None:
            print("DLL is empty can't delete!")
            return
        if self.head.nref is None:
            if x == self.head.data:
                self.head = None
            else:
                print("x is not present in the DLL!")
            return # for outer if condition
        
        if self.head.data == x: # when first node is the x
            self.head = self.head.nref
            self.head.pref = None
            return
        
        n = self.head
        while n.nref is not None:
            if x == n.data:
                break
            n = n.nref
        if n.nref is not None:# by this it will not check last node data because when n.ref
            # will be None then while loop will not execute 
            n.nref.pref = n.pref
            n.pref.nref = n.nref
            
        else:  # if control is here means n.nref is None i.e. it may be last node or node not
               # found in the doubly liked list
            if n.data == x:
                n.pref.nref = None
            else:
                print("x is not present in DLL")
                
                
                
                
                
                
        
            
dll1 = Doubly_LinkedList()
dll1.print_reverse()
# dll1.print_DLL()
```
