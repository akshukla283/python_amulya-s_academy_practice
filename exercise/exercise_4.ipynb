```python
### Exercise of QuickSort Algorithm

#  Ascending order.

# 1. Taking first element as a Pivot element 

# we will write QuickSort Algorithm in three steps 
# 1. Finding the Pivot Element Position
# 2. Divide left and right list from pivote and sort it recursively
# 3. main function for input and output



# 1. finding pivot place
def pivot_place(list1, first, last):  # here 'first' and 'last' is the index of the list
    
    pivot = list1[first]
    left = first + 1
    right = last
    
    while True:
            
        while left <= right and list1[left] <= pivot:
            left = left + 1

        while left <= right and list1[right] >= pivot:
            right = right - 1

        if right < left:
            break
            
        else:
            list1[left], list1[right] = list1[right], list1[left]
            
            
    list1[first], list1[right] = list1[right], list1[first]
    # here list1[left] is the pivot value changing with correct position at right 
    # because right < left now and we need to swap pivot value for it's correct place
        
    return right       

# 2. divide the list for the recursive call

def quicksort(list1, first, last):
    
    if first < last : # Base case # if single element is there then stop doing it
        
        p = pivot_place(list1, first, last)  # function call 
        # here p is the pivot index value
        
        # now divide the left part of pivot  by calling recursively
        quicksort(list1, first, p-1)
        quicksort(list1, p+1, last)
        
        
# 3. main
list1 = [56, 26, 93, 17, 31, 44]
n = len(list1)

quicksort(list1, 0, n-1)
# because index start from 0 in list1 and len(list1) will give 6 but we need 5
print(list1)

```

```python
### Exercise of QuickSort Algorithm

#  descending order.

# 1. Taking first element as a Pivot element 

# we will write QuickSort Algorithm in three steps 
# 1. Finding the Pivot Element Position
# 2. Divide left and right list from pivote and sort it recursively
# 3. main function for input and output



# 1. finding pivot place
def pivot_place(list1, first, last):  # here 'first' and 'last' is the index of the list
    
    pivot = list1[first]
    left = first + 1
    right = last
    
    while True:
            
        while left <= right and list1[left] >= pivot:  # for descending < to >
            left = left + 1

        while left <= right and list1[right] <= pivot:
            right = right - 1

        if right < left:
            break
            
        else:
            list1[left], list1[right] = list1[right], list1[left]
            
            
    list1[first], list1[right] = list1[right], list1[first]
    # here list1[left] is the pivot value changing with correct position at right 
    # because right < left now and we need to swap pivot value for it's correct place
        
    return right       

# 2. divide the list for the recursive call

def quicksort(list1, first, last):
    
    if first < last : # Base case # if single element is there then stop doing it
        
        p = pivot_place(list1, first, last)  # function call 
        # here p is the pivot index value
        
        # now divide the left part of pivot  by calling recursively
        quicksort(list1, first, p-1)
        quicksort(list1, p+1, last)
        
        
# 3. main
list1 = [56, 26, 93, 17, 31, 44]
n = len(list1)

quicksort(list1, 0, n-1)
# because index start from 0 in list1 and len(list1) will give 6 but we need 5
print(list1)

```
