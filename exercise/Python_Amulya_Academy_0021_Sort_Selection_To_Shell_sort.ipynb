### 1. Selection Sort lagorithm

```python
# 1. by using built-in method min and max for ascending and descending order respectively

list1 = [56, 2, 2, 0,3, 78, 0]
print("Unsorted : ",list1)
print()

for i in range(len(list1) -1):   # no need to sort the last element
    m_val = min(list1[i:])
    m_ind = list1.index(m_val, i)
    
    if list1[i] != m_val:
        list1[i], list1[m_ind] = list1[m_ind], list1[i]
        
    print(list1)
    
print()
print("Sorted List : ", list1)
```

```python
# for descending order 

list1 = [56, 2, 2, 0,3, 78, 0]
print("Unsorted : ",list1)
print()

for i in range(len(list1) -1):   # no need to sort the last element
    m_val = max(list1[i:])
    m_ind = list1.index(m_val, i)
    
    if list1[i] != m_val:
        list1[i], list1[m_ind] = list1[m_ind], list1[i]
        
    print(list1)
    
print()
print("Sorted List : ", list1)
```

### 2. Bubble Sort Algorithm

```python
list1 = [10, 5, 4, 23, 0]
print("Unsorted list: ", list1)
print()

for i in range(len(list1) - 1):
    for j in range(len(list1) - 1 - i):  # we took - i because we want only sorted step to print
        if list1[j] > list1[j+1]:
            list1[j], list1[j+1] = list1[j+1], list1[j]
            print(list1)
        else:
            print(list1)
            
    print() # need space on every iteration of outer for loop
    
print()
print("Sorted list : ", list1)
    
```

### 3. Quick Sort Algorithm 



```python
# first element as pivot element 
# For Ascending order

# 1. function for finding the currect pivot place

def pivot_place(list1, first, last):
    pivot = list1[first]
    left = first + 1
    right = last
    
    while True:
        # now we ned to right the condition for finding the currect position for 
        #the pivot element
        while left <= right and list1[left] <= pivot:
            left = left + 1
        while left <= right and list1[right] >= pivot:
            right = right -1

        if right < left :
            break
        else:
            # when if contidion is not satisfied  means left<= right and above two while loop
            # fails means we need to swap list left value  with list right value
            list1[left], list1[right] = list1[right], list1[left]
        # to check it until the if statement satified we neet to put it another while loop 
        # And when if contion will satisfied means 'break' statement execute and control will 
        # come out of while loop and then we need to swap list right value with pivot value 
        # which is the currect position of the while loop.
        
    list1[first], list1[right] = list1[right], list1[first]  # here list1[first] is pivot value
    
    # now we can return currect position of the pivot value
    return right 

# 2. funtion for quicsort after getting currection position of the pivot value

def quicksort(list1, first, last):
    # now contion to check whether there is a single value
    if first < last:
        p = pivot_place(list1, first, last)
        # by this we will get correct index of pivot value
        # now we can separte list into left and right list and called the function recursively
        # to sort the the list untill complete list sorted.
        quicksort(list1, first, p-1) # left list from pivot value
        quicksort(list1, p+1, last)  # right list from pivot value 
        
        
# 3. main 
list1 = [56, 26, 93, 17, 31, 44]
n = len(list1)
print("Before sorted : ", list1)
quicksort(list1, 0, n-1)  # here we took n-1 beacuse we need index not len of list

print()
print("After Sorted : ", list1)
        
```

```python
# Using Median of the list as pivot value
# ascending order 

# 1. function for finding the currect position of the pivot value.

import statistics  # for getting median of the list1

def pivot_place(list1, first, last):
    low = list1[first]
    high = list1[last]
    mid = (first + last)//2
    pivot_val = statistics.median([low, list1[mid], high])
    # now we need to check whether median value is low or high value
    if pivot_val == low:
        pindex = first
    elif pivot_val == high:
        pindex = last
    else:
        pindex = mid
    # now we will place pindex value at first index and then apply the algorithm
    list1[first], list1[pindex] = list1[first], list1[pindex]
    # now apply the algorithm 
    pivot = list1[first]
    left = first + 1
    right = last
    
    while True :
        while left<=right and list1[left] <= pivot:
            left = left + 1
        while left<=right and list1[right] >= pivot:
            right = right - 1
            
        if right < left :
            break
        else:
            list1[left], list1[right] = list1[right], list1[left]
            
    # swapping pivot value with right value  
    list1[first], list1[right] = list1[right], list1[first]
    
    return right 

# 2. function for quick sort

def quicksort(list1, first, last):
    if first < last:
        p = pivot_place(list1, first, last)
        
        quicksort(list1, first, p-1) # left list
        quicksort(list1, p+1, last) # right list
        
# 3. Main 
list1 = [56, 26, 93, 17, 31, 44]
n = len(list1)
print("Before sorted : ", list1)
quicksort(list1, 0, n-1)  # here we took n-1 beacuse we need index not len of list

print()
print("After Sorted : ", list1)
            
    
```

### 5. Insertion Sort Algorithm

```python

def insertionsort(my_list):
    
    for index in range(1, len(my_list)):
        current_element = my_list[index]
        pos = index
        while current_element < my_list[pos - 1] and pos > 0:
            my_list[pos] = my_list[pos-1]
            pos = pos - 1
            
        my_list[pos] = current_element
        
list1 = [2,4,3,5,1]
print("Unsorte list1: ",list1)
print()
insertionsort(list1)
print("Sorted list1: ", list1)
```

### 6. Shell Sort Algorithm

```python
def shellsort(alist):
    # first we need gap or incrmental value
    gap = len(alist)//2
    
    while gap > 0:
        for index in range(gap, len(alist)):  # we took range from gap because we sort by sublist
            # not by adjacent value
            current_element = alist[index]
            pos = index
            while pos>=gap and current_element < alist[pos - gap]:
                alist[pos] = alist[pos - gap]
                pos = pos - gap  # in insrtion operation we took -1 but here - gap because we
                # are comparing sublist.
                
            alist[pos] = current_element
            
        gap = gap//2
        
num = int(input("List Length: "))
list1 = [int(input(f"Enter {x+1} value in list: ")) for x in range(num)]
print("Unsorted list: ", list1)
print()
shellsort(list1)
print("Sorted list1: ", list1)
            
```
