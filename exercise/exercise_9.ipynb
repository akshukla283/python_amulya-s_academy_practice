```python
### Adding Deletion operations


# ------------------Node class ---------------------------

class Node :
    def __init__(self, data):
        self.data = data
        self.nref = None
        self.pref = None
        
# -------- Doubly LinkedList ------------------------------
class DoublyLL :
    def __init__(self):
        self.head = None
        
    #----- Traversal operation ----------------------------
    def print_LL(self):
        
        if self.head is None:
            print("LinkedList is empty!")
        else:
            n = self.head
            while n is not None:
                print(n.data, "--->", end = " ")
                n = n.nref
            
    def print_LL_reverse(self):
        print()  # to avoide same line
        if self.head is None:
            print("LinkedList is empty!")
            
        else:
            n = self.head
            while n.nref is not None:
                n = n.nref
            
            while n is not None:
                print(n.data, "--->", end = " ")
                n = n.pref
                
    # -------------insertion operation ----------------------
    def insert_empty(self, data):
        
        if self.head is None:
            new_node = Node(data)
            self.head = new_node
        else:
            print("LL is not empty!")
            
    def add_begin(self, data):
        
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            
        else:
            new_node.nref = self.head
            self.head.pref = new_node
            self.head = new_node
            
    def add_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            
        else:
            n = self.head
            while n.nref is not None:
                n = n.nref
            
            n.nref = new_node
            new_node.pref = n
            
    def add_after(self, data, x):
        if self.head is None:
            print("LL is empty!")
        else:
            n = self.head
            while n is not None:
                if x == n.data:
                    break
                n = n.nref
            
            if n is None:  # means x == n.data not satisfied and n become None
                
                print("Given node is not present in DLL!")
            else:  # if we are here means x == n.data is satisfied and we  are out of while loop
                new_node = Node(data)
                new_node.nref = n.nref
                new_node.pref = n
                if n.nref is not None:  # this is for checkking whether this is last node
                    n.nref.pref = new_node
                    
                n.nref = new_node
                
    def add_before(self, data, x):
        if self.head is None:
            print("DLL is empty!")
        else:
            n = self.head 
            while n is not None:
                if x == n.data:
                    break
                n = n.nref
            if n is None:
                print("Given node is not present in DLL!")
                
            else:
                new_node = Node(data)
                new_node.nref = n
                new_node.pref = n.pref
                if n.pref is not None:
                    n.pref.nref = new_node
                
                else:
                    self.head = new_node
                    
                n.pref = new_node
                
    # -------------- Deletion operation  ---------------------               
    def delete_begin(self):
        if self.head is None:
            print("DDL is empty can't delete!")
            return
        if self.head.nref is None:    # means here only one node
            self.head = None
            print("DLL is empty after deleting the node!")
            
        else:
            self.head = self.head.nref
            self.head.pref = None
            
    def delete_end(self):
        if self.head is None:
            print("DLL is empty we can't delete!")
            return
        if self.head.nref is None :    # means we have only one node
            self.head = None
            print("DLL is empty after deleting the node!")
            
        else:
            n = self.head 
            while n.nref is not None:
                n  = n.nref
                
            n.pref.nref = None
            
    def delete_by_value(self, x):
        if self.head is None:
            print("DLL is empty can't delete!")
            return 
        
        if self.head.nref is None:  # this is checking condition for having sinlge node
            if x == self.head.data:
                self.head = None
                
            else:
                print("x is not present in DLL!")
                return
            
        if self.head.data == x:  # when x is first node value
            self.head = self.head.nref
            self.head.pref = None
            return
            
        n = self.head
        while n.nref is not None:
            if x == n.data:
                break
            n = n.nref
            
        if n.nref is not None:
            n.nref.pref = n.pref
            n.pref.nref = n.nref
            
        else:
            if n.data == x : # means x is the last node 
                n.pref.nref = None
                return
            else:
                print("X is not present in DLL!")
                
            
            
                                                                                                                   

dll1 = DoublyLL()
dll1.add_begin(4)
dll1.add_before(10 , 4)
dll1.delete_by_value(10)
dll1.print_LL()
#dll1.print_LL_reverse()
```
