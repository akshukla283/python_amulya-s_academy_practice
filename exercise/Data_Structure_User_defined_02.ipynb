```python
### Adding element at end of the LinkedList

class Node:
    def __init__(self, data):
        self.data = data
        self.ref = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    #now traversal operation to print the data
    def print_LL(self):
        if self.head == None:
            print("Linked List is empty!")
        else:
            n = self.head
            while n is not None: # we took n because we want check whether data is present
                print(n.data,"--->", end = " ")
                n = n.ref
                
    def add_begin(self, data): # we need extra parameter as data or element which to be add
        new_node = Node(data)
        new_node.ref = self.head # storing the present head value as the reference in new_node
        self.head = new_node  # and making new_node as head of the Linkedlist
        
    def add_end(self, data):
        # first check whether linked list is empty if empty then just create a node 
        # if it is not empty then we need to at last node and add new node
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            n = self.head
            while n.ref is not None: # here we took n.ref becasue we know at end of the linkedList
                # will have no reference when this loop we get that then control will come-out
                # of this.
                n = n.ref
                
            # when control will be here means we are at last node now we can add the new node
            n.ref = new_node
        
                
LL = LinkedList()
LL.add_begin(20)
LL.add_begin(30)
LL.add_begin(40)
LL.add_end(0)
LL.add_end(4)

LL.print_LL()
```

```python
#### adding element after x element, where x is the part of the LinkedList

class Node:
    def __init__(self, data):
        self.data = data
        self.ref = None
        
class LinkedList:
    def __init__(self):
        self.head = None  ## initially head should none means no Node is present
        
    # traversal operation to print the data
    def print_LL(self):
        if self.head is None:
            print("LinkedLIst is empty!")
            
        else:
            n = self.head
            while n is not None:  # print the data till the condition true.
                print(n.data, "--->", end = " ")
                n = n.ref
    # adding element at beginning to no need to check whether list is empty or not just add 
    def add_begin(self, data):
        new_node = Node(data)
        new_node.ref = self.head
        self.head = new_node
    
    # add data at the end for this first we need to check whether LinkedList is empty or not 
    # if empty then just add data as beginning if present then travel to last element and add
    def add_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            n = self.head
            while n.ref is not None:
                n = n.ref
                
            n.ref = new_node
            
    def add_after(self, data, x):
        # first we need to travel through the all element to check whether x is present there
        # if present then add else print message that x is not present
        n = self.head 
        while n is not None:
            if n.data == x:
                break
            n = n.ref
        # control will be here by two reason 1. when loop is terminated by 'break' statement
        # and 2. when n become 'None' means even after traveling through all the element we didn't
        # find the 'x' after which we wanted to add the new element
        if n is None:
            print("Node is not present in the Linked list")
        else:
            new_node = Node(data)
            new_node.ref = n.ref
            n.ref = new_node
            
            
        
LL = LinkedList()
LL.add_begin(20)
LL.add_begin(12)
LL.add_begin(90)
LL.add_after(100, 12)
LL.print_LL()
```
