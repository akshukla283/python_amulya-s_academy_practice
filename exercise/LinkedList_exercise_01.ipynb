```python
### Singly Linked List
# in singly linked list whenever we will create new node that have head , data, reference None
#   ________node____
#  |__data__|__ref_| here ref = none
```

```python
# class for making new node
class Node:
    def __init__(self, data):
        self.data = data
        self.ref = None
        
# class for Linked List
class LinkedList:
    def __init__(self):
        self.head = None
        
    # printing the data present in linkedlist
    def print_LL(self):
        if self.head is None:
            print("Linked List is empty!")
        else:
            n = self.head
            while n is not None:
                print(n.data, "--->", end = " ")
                n = n.ref
    # add new node at beginning
    def add_begin(self, data):
        new_node = Node(data)
        new_node.ref = self.head
        self.head = new_node
        
    # adding element at end of the linked list
    def add_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            n = self.head 
            while n.ref is not None:
                n = n.ref
                
            n.ref = new_node
                
            
            
    # adding element after a particular element
    def add_after(self, data, x):
        n = self.head
        while n is not None:
            if n.data == x:
                break
            else:
                n = n.ref
        # control will be here by 2 reason only first either by break statement or when n == None
        if n is None:
            print("Node is not present in the Linked List!")
        else:
            new_node = Node(data)
            new_node.ref = n.ref
            n.ref = new_node
    # adding an element before a particular element
    def add_before(self, data, x):
        
        # when LinkedList is empty
        if self.head is None:
            print("Linked List is empty so can't find the node")
            return
        # when first node is that particular node before we will add new element
        if self.head.data == x:
            new_node = Node(data)
            new_node.ref = self.head
            self.head = new_node
            return
        
        # now we need to travel all the node to find the particular node
        n = self.head 
        while n.ref is not None:
            if n.ref.data == x:
                break
            else:
                n = n.ref
        if n.ref is None:
            print("Node is not present in the Linked List!")
        else:
            new_node = Node(data)
            new_node.ref = n.ref
            n.ref = new_node
            
    # add new element when node is empty
    def add_empty(self, data):
        if self.head is None:
            new_node = Node(data)
            self.head = new_node
        else:
            print("Linked List is not empty!")
            
            
ll1 = LinkedList()
#ll1.add_empty(10)
# ll1.add_empty(10)

# ll1.add_begin(10)
# ll1.add_begin(20)
# ll1.add_begin(30)

# ll1.add_end(10)
# ll1.add_end(30)
# ll1.add_end(40)

# #ll1.add_before(10, 20)
# ll1.add_before(20,10)
# ll1.add_before(100,10)


# ll1.add_after(20,10)
# ll1.add_after(300,20)

#ll1.add_begin(500)



ll1.print_LL()
```
