#### Sleep method


```python
import time
print("Hello people")

# we will get output instantly because we haven't set sleep time now in next program we will set it 
```

```python
import time

time.sleep(5) # we set time 5 second
print("Hello people")
```

```python
import time
# using time.sleep method we can delay the execution of time
lst = [1,2,3,4,5]

for i in lst:
    time.sleep(2)
    print(i)
```
