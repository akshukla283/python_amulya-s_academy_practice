### Doubly Linked List | Data Structure

#### Doubly Linked List
`We add a pointer to the previous node in a doubly-linked list. Thus, we can go in either direction: forward or backward.`

```python
# In Doubly Linked list there will be pre-reference and next reference and a head

#   ________Node_______
#  | pref | head |nref |
#  |______|______|_____|
```
