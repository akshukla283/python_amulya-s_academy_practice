### Delete Any Node By Value in Linked List | Delete at the end

```python

class Node:
    def __init__(self, data):
        self.data = data
        self.ref = None

class LinkedList:
    def __init__(self):
        self.head = None  # means we have no node empty LinkedList
        
    def print_LL(self):     # Traversal operations
        if self.head is None:
            print("LinkedList is empty!")
        else:
            n = self.head
            while n is not None:
                print(n.data,"--->", end = " ")
                n = n.ref
                
    def add_begin(self,data):
        new_node = Node(data)
        new_node.ref = self.head
        self.head = new_node
        
    def add_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            n = self.head
            while n.ref is not None:
                n = n.ref
                
            n.ref = new_node
    def add_after(self, data, x):
        n = self.head
        while n is not None:
            if x == n.data:
                break
            n = n.ref
        if n is None:
            print("node is not present in LL")
        else:
            new_node = Node(data)
            new_node.ref = n.ref
            n.ref = new_node
    def add_before(self, data, x):# we need to check whether linkedlist is empty, because empty linked
                                   # will give error so avoidthis we will use condition 
        if self.head is None:
            print("LinkedList is empty!")
            return
        
        if self.head.data == x:
            new_node = Node(data)
            new_node.ref = self.head
            self.head = new_node
            return
        n = self.head
        while n.ref is not None:
            if n.ref.data == x:
                break
            n = n.ref
        if n.ref is None:
            print("Node is not found!")
        else:
            new_node = Node(data)
            new_node.ref = self.head
            self.head = new_node
            
    def insert_empty(self, data):
        if self.head is None:
            new_node = Node(data)
            self.head = new_node
        else:
            print("LinkedList is not empty!")
            
    def delete_begin(self):
        if self.head is None:
            print("LinkedList is empty so we can't delete nodes!")
            
        else:
            self.head = self.head.ref
    
    def delete_end(self):
        if self.head is None:
            print("LL is empty so we can't delete nodes!")
        else:
            n = self.head
            while n.ref.ref is not None:
                n = n.ref
            n.ref = None
    def delete_by_value(self, x):
        if self.head is None:
            print("can't deelete LL is empty!")
            return # we use return so that after printig control should come out of the loop.
        # now chaeck wether the node is first node or not 
        if x == self.head.data:
            self.head = self.head.ref
            return  # if we don't want write else part then write 'return' or elif elif esle
        n = self.head  # using this we will search the head of the previous node
        while n.ref is not None:
            if x == n.ref.data:
                break
            n = n.ref
        #now we are here by two condition one is when while loop satisfied and second is when
        # nested if inside while loop satisfied and 'break' statement works.
        if n.ref is None:
            print("Node is not Found!")
        else:
            n.ref = n.ref.ref
            
            

LL1 = LinkedList()
LL1.add_begin(10)
LL1.add_begin(20)
LL1.add_begin(30)
LL1.delete_by_value(10)
LL1.print_LL()
```
